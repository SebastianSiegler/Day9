```python
# 1 . Read the winequality-red.csv data into a DataFrame, and the winequality-white.csv into
# another DataFrame.

import pandas as pd
import numpy as np

red_wines_df = pd.read_csv('data/winequality-red.csv', delimiter=';')
white_wines_df = pd.read_csv('data/winequality-white.csv', delimiter=';')

red_wines_df.head()
```

```python
# 2. Double check that you've read them in right by using some of the attributes and methods available on
# DataFrames for getting a general sense of your data (Hint: See day10-intro_pandas, week5
# if you need a refresher on these).

red_wines_df.head()
white_wines_df.tail()
```

```python
# 3.  I've decided that this month I want to stay away from wines with relatively high alcohol content.
# To do that, I'm going to avoid any wines that have a greater alcohol content than the mean alcohol content,
# and you're going to help me do this. To achieve this, let's do the following:

# Find the mean alcohol content, separately, for reds and whites.

red_wine_alcohol_mean = red_wines_df.alcohol.mean()
red_wine_alcohol_mean

white_wine_alcohol_mean = white_wines_df.alcohol.mean()
white_wine_alcohol_mean

# Create a Series that holds whether each row in each DataFrame (the reds, and whites) has a higher
# alcohol content than the mean.

check_for_alcohol_mean_red = red_wines_df["alcohol"] > red_wine_alcohol_mean
check_for_alcohol_mean_red

check_for_alcohol_mean_white = white_wines_df["alcohol"] > white_wine_alcohol_mean
check_for_alcohol_mean_white

# Merge this Series onto the DataFrame. I can imagine doing this with either a .join() or using pd.concat().
# For practice, do it with both. Note: merges with Series work the same way that they work with DataFrames.


check_for_alcohol_mean_red.name = "alcohol_bool_red"
check_for_alcohol_mean_white.name = "alcohol_bool_white"

red_wines_joined = red_wines_df.join(check_for_alcohol_mean_red)
white_wines_joined = white_wines_df.join(check_for_alcohol_mean_white)

# Return back to me all those rows that will help me stay away from those wines with a higher alcohol content.

red_wines_joined[red_wines_joined['alcohol_bool_red'] == False]


```

```python
# 4. Let's say that I want to get started on cutting back next month. This time, though, I want to focus on
# staying away from those wines with a high acidity. Specifically, I want to stay away from those wines
# that are in the highest bin of fixed acidity (highest bin out of 5). You're now going to help me do this.
# To achieve this, let's do the following:

# Separate the rows in each DataFrame into 5 equal width bins based off their fixed acidity.

red_wines_df.columns = [col.replace(' ', '_') for col in red_wines_df.columns.tolist()]
white_wines_df.columns = [col.replace(' ', '_') for col in white_wines_df.columns.tolist()]

fixed_acidity_cut = pd.cut(red_wines_df.fixed_acidity, bins = 5, labels=[1,2,3,4,5])
fixed_acidity_cut.name = "fixed_acidity_bin"


fixed_acidity_cut_white = pd.cut(white_wines_df.fixed_acidity, bins = 5, labels=[1,2,3,4,5])
fixed_acidity_cut_white.name = "fixed_acidity_bin"



# Merge the resulting Series holding these 5 bins onto the original DataFrame. I can imagine also doing
# this with either .join() or using pd.concat(). Try doing it with both for practice.


red_wines_joined = red_wines_df.join(fixed_acidity_cut)
red_wines_joined

white_wines_joined = white_wines_df.join(fixed_acidity_cut_white)
#white_wines_joined

# Return back to me all those rows that are *not in * the top bin in terms of fixed acidity.

red_wines_joined[red_wines_joined.fixed_acidity_bin < 5]
white_wines_joined[white_wines_joined.fixed_acidity_bin < 5]
```

```python
# 5 ) Let's say that I now want to know how much my decision to avoid those wines with higher alcohol content
# is going to limit the quality of wines that I can drink. To figure this out, I want to know a couple
# of things:

# The average alcohol content for those reds above the mean alcohol level, by quality.

pd.pivot_table(red_wines_df[red_wines_df.alcohol > red_wines_df.alcohol.mean()], values="alcohol", index="quality")

# Find the average alcohol content for whites, by quality and fixed acidity quintile.

pd.pivot_table(white_wines_df[white_wines_df.alcohol.mean()])

```
